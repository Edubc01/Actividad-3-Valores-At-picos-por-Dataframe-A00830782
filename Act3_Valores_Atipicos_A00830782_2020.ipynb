Convertimos algunas variables que son strings a flotantes para poder trabajar con dichos datos.

```python
df['RET ISR'] = df['RET ISR'].astype('float')
df['RET IVA'] = df['RET IVA'].astype('float')
df['TOTAL MX'] = df['TOTAL MX'].astype('float')
df['TOTAL SAT'] = df['TOTAL SAT'].astype('float')
```

```python
df.dtypes
```

Creamos un nuevo dataframe con algunas de las variables del dataframe pasado para sacar los valores atipicos, y poder eliminar outliers con metodos de desviacion estandar y rango intercuartilico.

```python
df1 = df[['TC', 'IMPORTE', 'IVA', 'RET ISR', 'RET IVA', 'TOTAL MX', 'TOTAL SAT']]
df1
```

Desviacion estandar para eliminar outliers en DataFrame nuevo.

```python
y= df1
Limite_Superior= y.mean() + 3 * y.std()
Limite_Inferior= y.mean() - 3 * y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
df2 = df1[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df2
```

```python
valores_nulos = df1.isnull().sum()
valores_nulos
```

```python
df_limpio=df2.copy()
df_limpio=df_limpio.fillna(round(df2.mean(),1))
df_limpio
```

```python
valores_nulos=df_limpio.isnull().sum()
valores_nulos
```

```python
df_limpio.to_csv("DF 2020 1, Desviacion estandar.csv")
```

Rango intercuartilico para eliminar outliers en DataFrame nuevo.

```python
y=df1

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
df3_iqr= df1[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df3_iqr
```

```python
valores_nulos=df3_iqr.isnull().sum()
valores_nulos
```

```python
df5_iqr=df3_iqr.copy()
df5_iqr=df5_iqr.fillna(round(df3_iqr.mean(),1))
df5_iqr
```

```python
df5_iqr.to_csv("DF 2020 2, Cuartiles.csv")
```
